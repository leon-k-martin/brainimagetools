# Chord Diagram

```python
import numpy as np
import mne
import brainvistools
from os.path import join, abspath, dirname
from mne.viz import circular_layout
from mne_connectivity.viz import plot_connectivity_circle
import matplotlib.pyplot as plt

labels = []
node_colors = []
unknown_found = None

with open(join(dirname(brainvistools.__file__), "data", "FreeSurferColorLut.txt")) as f:
    for x in f.readlines():
        if x.startswith("#") or x == "\n":
            continue

        if unknown_found and x.split():
            labels.append(x.split()[2])
            node_colors.append(tuple([float(color) / 255 for color in x.split()[3:6]]))
        if "Unknown" in x:
            unknown_found = True

n_nodes = 10
node_angles = circular_layout(list(set(labels))[:n_nodes], list(set(labels))[:n_nodes])

connectome = np.random.random((n_nodes, n_nodes))


fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection="polar")
ax1.imshow(connectome)
plot_connectivity_circle(
    connectome,
    labels[:n_nodes],
    node_angles=node_angles,
    node_colors=node_colors,
    ax=ax2,
    colorbar=False,
    fig=fig,
    interactive=False,
    facecolor="white",
    padding=0,
    textcolor="black",
)
```
